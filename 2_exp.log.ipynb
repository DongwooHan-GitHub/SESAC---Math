```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 100)
y1 = 2**(x-1)
y2 = 2**(x-1) +1

# plt.plot(x, y, label = "2^x") 이건 제가 주석처리 해놓았어요! 실행시켜보려구 ㅎㅎ
plt.plot(x, y1, label = "2^(x-1)")
plt.plot(x, y2, label = "2^(x-1) +1 ", linestyle = ":")

plt.grid(linestyle = ":")
plt.vlines(0, 0, 16, color = "b")
plt.legend()
plt.show()
```
