```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polynomial(x, n):
    y = x**n
    return y

x = np.linspace(-4, 4, 100)
# print(x)
y = polynomial(x,1) #linear fun
y2 = polynomial(x,2)
y3 = polynomial(x,3)

plt.plot(x, y, label = "y = x")
plt.plot(x, y2, label = "y = x^2")
plt.plot(x, y3, label = "y = x^3")
plt.plot(x,polynomial(x,4), label = "y = x^4")

plt.title("polymomial curve")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.legend()
plt.ylim(-2, 2)
plt.xlim(-2, 2)

plt.grid(linestyle = ":")
plt.show()
```

```python

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)

#   
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

y_list = [sin_x,cos_x,tan_x]
y_label = ["sin(x)","cos(x)","tan(x)"]
color_list = ["b", "g", "r"]
fig, ax = plt.subplots(3,1, figsize = (12, 10))
for i in range(3):
    ax[i].plot(x,y_list[i], label = y_label[i], c = color_list[i])
    ax[i].set_ylim(-1, 1)
    ax[i].grid(linestyle = ":")
    ax[i].set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi,", "pi", "0","pi","2pi"])

plt.show()    
```
